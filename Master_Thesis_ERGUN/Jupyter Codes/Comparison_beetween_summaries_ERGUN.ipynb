```python
Rouge1 =[]
Rouge2 = []
for index,row in TOFO.iterrows():

  from sacrerouge.metrics import Rouge
  rouge = Rouge(max_ngram=2)
  k = rouge.score(row['T5_pretrained_Summary'], [row['Human Summary']])
  k1 = rouge.score(row['T5_finetuned_Summary'], [row['Human Summary']])


  print("\nDistance between the summaries ppretrained : ",k['rouge-1']['f1'])
  print("\nDistance between the summaries finetuned : ",k1['rouge-1']['f1'])
  
  print("-"*200)
  Rouge1.append(k['rouge-1']['f1'])
  Rouge2.append(k1['rouge-1']['f1'])
```

```python
TOFO["Rouge_pre"]= Rouge1
TOFO["Rouge_fin"]= Rouge2
```

```python
TOFO = TOFO[['Human Summary', "T5_pretrained_Summary", "WMD_pre","Rouge_pre", "T5_finetuned_Summary", "WMD_fin","Rouge_fin"]]
TOFO
```
